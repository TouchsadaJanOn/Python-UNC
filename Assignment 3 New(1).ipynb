# Algorithm 1 - Tortoise Practicing
The tortoise is practicing for the upcoming race with the hare. This algorithm traces prgress of the tortoise, given his speed and the distance of the race.


```python
print("Tortoise Alone 1")
speed = int(input("speed = "))
raceDistance = int(input("distance = "))
print()
print("%3s %15s"%("Time","Current"))
Time = 0
Current = 0
while Current <= raceDistance or Current <= raceDistance + 4:
    print("%3d %12d" %(Time,Current))
    Time += 1
    Current += speed
print("The race is over!")
```

# Algorithm 2 - Hare Practicing
The hare is practicing for the upcoming race with the tortoise. This algorithm traces progress of the hare, given his speed and the distance of the race. Also provided is the start and the duration of the nap.

```python
print("Hare Alone")
speed = int(input("speed = "))
distance = int(input("distance = "))
napStart = int(input("napStart = "))-1
napDuration = int(input("napDuration = "))
print()
print("%4s %15s %15s"%("Time","Current", "Naps?"))
time = 0
current = 0
napCount = 0
print("%3d %13d" % (time, current))
time += 1
while current <= distance:
    #Naps
    if time > 0 and napStart < time <= napStart + napDuration:
        napCount += 1
        print("%3d %13d %23s" % (time, current, "Napping "+ str(napCount)))
    #Not Napping
    else:
        current += speed
        print("%3d %13d" % (time, current))       
    time += 1
print("The race is over!")
```

# Algorithm 3 - Tortoise vs. Hare
The race is on! The tortoise and the hare are ready to race. This algorithm traces progress of the tortiose and hare, given the distance of the race, their respective speeds, and the nap start time and duration for the time.

```python
print("Tortoise v. Hare program\nBy Touchsada Jan On, CS120")
print()
print("This program recreates the famed story, The Tortoise and the Hare. Please enter the input values below.")
print()
#User input here
racedistance = int(input(f"\tRace distance (feet): "))
harespeed = float(input(f"\tHare's speed (feet/sec): "))
tortoisespeed = float(input(f"\tTortoise's speed (feet/sec): "))
harenapstart = int(input(f"\tNap start time: "))
harenapduration = int(input(f"\tNap duration: "))
print()
print("Thank you. Here is your race.")
print("----------------------------------------------------------------------------------------------------------")
def race(racedistance, tortoisespeed, harespeed,harenapstart, harenapduration):
    time = 0
    tortoisecurrent = 0
    harecurrent = 0
    napcount = 1
    # Condition for the distance to go one more loop so it can print the last distance
    tortwin = False
    harewin = False
    # If the animal are tie
    both = harecurrent==tortoisecurrent
    both = False
    hareoriginal = harespeed
    comment = ""
    print("%30s %30s" % ("Tortoise","Hare"))
    print("%4s %20s %10s %21s %10s %15s"%("Time","Position","Speed","Position","Speed","Comments"))
    
    for time in range(1000000):
        print("%4d %20s %10s %20s %11s %15s"%(time,"{0:0.1f}".format(tortoisecurrent), tortoisespeed,
                                            "{0:0.1f}".format(harecurrent), harespeed, comment))
        tortoisecurrent += tortoisespeed
        harecurrent += harespeed
        #Naps
        if harenapstart <= time <= harenapstart + harenapduration-1:
            harespeed = 0
            comment = "\t\tNapping second " + str(napcount)
            napcount += 1
        # No Naps
        else:
            harespeed = hareoriginal
            comment = ""
        # Condition will be True if one of the animal have their position >= racedistance)
        # If one of the animal won(True), will go through the loop one more time and print the statement)
        if tortwin:
          print("\t\t\t\t\t\t\t\t\t Race over - Tortoise Wins")
          break
        if harewin:
          print("\t\t\t\t\t\t\t\t\t Race over - Hare Wins")
          break
        if both:
          print("\t\t\t\t\t\t\t\t\t Race over - It's a Tie")
          break
        if tortoisecurrent > racedistance:
            tortwin = True
        elif harecurrent > racedistance:
            harewin = True
        #If the animals are tie
        elif harecurrent == tortoisecurrent and harecurrent == racedistance:
            both = True
        
race(racedistance, tortoisespeed, harespeed,harenapstart, harenapduration)
```

# Algorithm 4 - The Big Race
The horse is practicing for the upcoming race with the tortoise and the hare. This algorithm traces progress of the horse, given his speed and the distance of the race. Also provided is the start and the duration of the horse stopping and drinking some water.

```python
print("The Big Race \nBy Touchsada Jan On, CS120")
print()
#User input
racedistance = int(input(f"\tRace distance (feet): "))
harespeed = float(input(f"\tHare's speed (feet/sec): "))
tortoisespeed = float(input(f"\tTortoise's speed (feet/sec): "))
horsespeed = float(input(f"\tHorse's speed (feet/sec): "))
harenapstart = int(input(f"\tNap start time: "))
harenapduration = int(input(f"\tNap duration: "))
horsefatigueStart = int(input(f"\tHorse fatique start time: "))
horsefatigueDuration = int(input(f"\tHorse fatique duration: "))
print()
print("Thank you. Your race is in TheBigRace file.")
print("----------------------------------------------------------------------------------------------------------")

def race(racedistance, tortoisespeed, harespeed, horsespeed, harenapstart, harenapduration):
    #Write to file
    TheBigRace = open("TheBigRace.txt", "w")
    TheBigRace.write("The Big Race \nBy Touchsada Jan On, CS120\n")
    #Animal Title
    print("%20s %20s %20s" % ("Tortoise","Hare", "Horse"), file =TheBigRace)
    time = 0
    tortoisecurrent = 0
    harecurrent = 0
    horsecurrent = 0
    napcount = 1
    tortwin = False
    harewin = False
    horsewin = False
    both = harecurrent==tortoisecurrent
    both = False
    horseoriginal = horsespeed
    hareoriginal = harespeed
    print("%4s %8s %10s %10s %10s %10s %10s %15s" %("Time","Position","Speed","Position",
                                                     "Speed","Position","Speed","Comments"),file= TheBigRace)
    #Comment will include hare napping and if the horse get tired once it goes through the loop
    comment = ""
    for time in range(1000000):
        print("%4d %8d %10d %10d %10d %10d %10d %15s" %(time,tortoisecurrent,tortoisespeed,harecurrent,
                                                  harespeed,horsecurrent,horsespeed,comment),file = TheBigRace)
       #Each time speed will add to current(position) of each animal
        tortoisecurrent += tortoisespeed
        harecurrent += harespeed
        horsecurrent += horsespeed
        #Nap
        if harenapstart <= time <= harenapstart + harenapduration-1:
            harespeed = 0
            napcount += 1
            comment =  "\t\tHare Napping " + str(napcount)
       #Horse fatigue
        elif horsefatigueStart <= time <= horsefatigueStart + horsefatigueDuration:
            horsespeed -= 2
            comment =  "\t\tHorse getting tired"
           
        else:
            harespeed = hareoriginal
            comment = ""
        if horsewin:
            TheBigRace.write("\t\t\t\t\t\t\t\t\t Race over - Horse Wins")
       
            break
        if tortwin:
            TheBigRace.write("\t\t\t\t\t\t\t\t\t Race over - Tortoise Wins")
           
            break
        if harewin:
            TheBigRace.write("\t\t\t\t\t\t\t\t\t Race over - Hare Wins")
            break
        if tortoisecurrent > racedistance:
            tortwin = True
        if harecurrent > racedistance:
            harewin = True
        if horsecurrent >= racedistance:
            horsewin = True
    TheBigRace.close()
       
race(racedistance, tortoisespeed, harespeed,horsespeed, harenapstart, harenapduration)
```
