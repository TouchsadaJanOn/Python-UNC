```python
import random
from random import randrange

def Greeting():
    print("This program will select a phase depending on the temperature of the weather chosen in Celsius! Let's Do It!\n")
    
    #Produce random number
    day = randrange(1,31)
    month = randrange(1,13)
    year = randrange(2000,2050)
    print("Hypothetically, Today is!")
    
    #string format
    print("%4s %10s %10s" %("Day", "Month", "Year"))
    print("%4s %10s %10s" %(day, month, year))
    print("\nFirst, Let's get to know you!!\n")
    name = input("\nPlease enter your name here: ")
    print("\nGreetings " + name +"!")
    mood = str(input("How are you feeling today, good? Y/N: "))
    if mood[0] == "G" or mood[0] == "g" or mood[0] == "Y" or mood[0] == "y":
        print("\nGreat! Let's keep going")  
    else:
        print("\nAwww! It will be OK. Let's keep going!")



def converting():
    cel = int(input("\nThe Celsius Temperature: "))
    fahr = 9/5 * cel +32
    print("Here is the Fahrenheit temperature: ", fahr)
    print("\nHope that helps! Let's move on!")
    
def deciding():
    con = input("\nWould you like to convert Celsius to Fahrenheit? Y/N: ")
    if con[0] == "Y" or con[0] == "y":
        converting()
    else:
        print("\nAwesome! Let's move on!")

def phase():
    outfile = open("Temperature History.txt","w")
    
    cold1 = ["Bhrrrrr, it's cold!","The weather is definitely colder than my heart!","Hello! Frostbites...",
            "Time to book a flight to Florida!","Time to grow out all the hair for extra warmth.",
             "Everyday, I'm shovelling","Frostbite or Icetan?","Dang it's cold! My farts are snowflakes!",
             "Breakfast! Frosted Flakes!","Forcast: There will be a snowstorm tomorrow, School: Don't miss class tomorrow."]
    mid1 = ["Please Mother Nature! Make up your mind! Cold or Warm!","Be average...",
            "Hope you know how to dress, cuz I don't!","Got Gloves? Meh.. you don't need it!",
            "Don't forget hot water! Get the ice off the windshield!",
            "Shave half of your body. It might get hot later...",
            "Bring sunglasses to hide your tears, it'll be windy.",
            "Snow and Dirt! Black and whilte, maybe yellow!",
            "Two in One! Get some sunburn and frostbite!","It's so cold. It's so hot. Hmm I'm sick..."]
    hot1 = ["Ice cream! Melted Cream!","Hello Lobster, you're a little burn!",
            "First person to complain will get punched!","Let's fry an egg with the Sun",
            "My body is on fire!!","Yo Weather! Stop showing off, we know you're hot!",
            "Why are there two of you?","Introducing my new home: Fridge.",
            "I'm so Hot! Why am I not in a relationship?","Too hot to even move! I'll do everything later..."]
    num = int(input("\nHow many temperature inputs are you planning: "))
    history = []

    #For Loop
    for i in range(num):
        user = input("\nWhat is temperature in Celsius you like: ")
        history.append(float(user))  
        if int(user) <= -38:
            print((cold1[random.randint(0, len(cold1)-1)]+" ")*2+"\nSorry for the stutter! It is sssooo ccolld.")
            print(((cold1[random.randint(0, len(cold1)-1)]+" ")*2)+"\nSorry for the stutter! It is sssooo ccolld.",
                  file=outfile)
        if int(user) <= 0 and int(user) >= -37:
            print(cold1[random.randint(0, len(cold1)-1)])
            print(cold1[random.randint(0, len(cold1)-1)],file=outfile)
        if int(user) > 0 and int(user) <= 27:
            print(mid1[random.randint(0, len(mid1)-1)])
            print(mid1[random.randint(0, len(mid1)-1)],file=outfile)
        if int(user) >= 28 and int(user) <=37:
            print(hot1[random.randint(0, len(hot1)-1)])
            print(hot1[random.randint(0, len(hot1)-1)],file=outfile)
        if int(user) >= 38:
            print(hot1[random.randint(0, len(hot1)-1)].upper())
            print(hot1[random.randint(0, len(hot1)-1)].upper(),file=outfile)
    his = input("\nWould you like to see your temperature history? Y/N: ")
    if his[0] == "N" or his[0] == "n":
        print("\nGreat! We start over and Let's keep going")
        phase()
    else:
        print("\nCool! Here it is! You can also find it in the file titled 'Temperature History'!")
        print(history)
        print(history,file=outfile)
```

```python
Greeting()
deciding()
phase()
```
