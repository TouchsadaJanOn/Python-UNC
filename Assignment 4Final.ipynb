# AnimalClass
This class will represent one participant or contestant in the race.

```python
class Animal:

   
    def __init__(self,name,species,speed,napStart,napDuration,currentPosition):
        self.name = name
        self.species = species
        self.speed = float(speed)
        self.napstart = napStart
        self.napduration = napDuration
        self.currentposition = float(currentPosition)
        self.time = 0
        self.napping = False
        return
       
    # get method
    def getName(self):
        return self.name
    def getSpeed(self):
        return self.speed
    def NapStart(self):
        return self.napstart
    def getNapDuration(self):
        return self.napduration
    def getCurrentPosition(self):
        return self.currentposition
    def getSpecies(self):
        return self.species
    # set method
    def setSpeed(self, num):
        self.speed = num
        return
    def setNapStart(self, num):
        self.napstart = num
        return
    def setNapDuration(self, num):
        self.napduration = num
        return
    def setCurrentPosition(self, num):
        self.napCurrentPos = num
        return
    def advancePosition(self):
        
        if self.time >= 0 and self.napstart < self.time and self.time < self.napstart + self.napduration:
            self.time += 1
            self.napping = False
        else:
            self.time += 1
            self.currentposition += self.speed
            self.napping = True
    def __str__(self):
        return str.format("%10s %8s" % ("{0:0.1f}".format(self.currentposition),"{0:0.1f}".format(self.speed)))
```

# Class Race
This class will represent two participant or contestants in a single race, also include default event name and race distance.


```python
class Race:
    def __init__(self, contestant1, contestant2,eventname="Default Race Name", distance =100):
        self.eventname = eventname
        self.distance = distance
        self.con1 = contestant1
        self.con2 = contestant2
        self.time = 0
        self.napcount = 0
    def getEventName(self):
        return self.eventname
    def getDistance(self):
        return self.distance
    def runRace(self):
        comment = ""
        original = self.con2.speed
        print("%4s %10s %8s %10s %8s" %(self.time,self.con1.currentposition, self.con1.speed,self.con2.currentposition,self.con2.speed))
        while (self.con1.getCurrentPosition() < self.distance and self.con2.getCurrentPosition() < self.distance):
            comment = ""
            #Position
            self.con1.advancePosition()
            self.con2.advancePosition()
            #Defining animal base on race object and contestant object
            #Tortoise
            con1 = Animal(self.con1.name, self.con1.species, self.con1.speed,self.con1.napstart,self.con1.napduration,self.con1.currentposition)
            #Hare
            con2 = Animal(self.con2.name, self.con2.species, self.con2.speed,self.con2.napstart,self.con2.napduration,self.con2.currentposition)
        
            self.time += 1 
            
            if self.time == self.con2.napstart:
                comment = "\tNap Started"
            else:
                if self.time == self.con2.napstart+self.con2.napduration:
                    comment = "\tNap second " +str(self.napcount+1) + ", ending"
            
              #Naps
            if (self.con2.napstart < self.time < self.con2.napstart + self.con2.napduration):
                self.napcount += 1
                self.con2.speed = 0
                comment = "\tNapping second " + str(self.napcount)
      
            else:
                self.con2.speed = original
      
            #Hare Wins      
            if self.con1.getCurrentPosition() > self.con2.getCurrentPosition() and self.con1.getCurrentPosition() >= self.distance:
                comment = "\tRace over - " + con1.species + " Wins"
            #Tortoise Wins        
            if (self.con2.getCurrentPosition() > self.con1.getCurrentPosition() and self.con2.getCurrentPosition() >= self.distance):
                comment = "\tRace over - " + con2.species + " Wins"
            #Both are Tied
            if (self.con1.getCurrentPosition() == self.con2.getCurrentPosition() and self.con1.getCurrentPosition() >= self.distance):
                comment = "\tRace over - " + con1.species +" and " + con2.species + " are Tied!"
        
            
            print("%4d %4s %4s %4s" % (self.time, self.con1, self.con2, comment))
            
    def __str__(self):
        return str.format("%0s %12s %20s %15s" % (self.con1, self.con2, self.eventname, self.distance))
```
